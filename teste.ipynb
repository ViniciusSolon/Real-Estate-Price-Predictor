```python
import pandas as pd

df = pd.read_csv("area-preco.csv")
df
```

```python
df.plot(kind="scatter", x="Area", y="Preco");
```

```python
from sklearn import linear_model

modelo = linear_model.LinearRegression()
x = df[["Area"]]
y = df[["Preco"]]

modelo.fit(x, y)
```

```python
print(modelo.predict([[15]])[0][0])
```
